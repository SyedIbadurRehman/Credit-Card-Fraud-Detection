Importing the libraries

```python
import numpy as np
import pandas as pd
from pandas.core.base import value_counts
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MaxAbsScaler
```

Reading the files

```python
df=pd.read_csv("creditcard.csv", error_bad_lines=False)
```

```python
df.head()
```

Statistical Measure of Data

```python
df.describe().T
```

```python
df.shape
```

Finding the Target Output counts

Turns out the data is heavily UNDER SAMPLED

```python
df.Class.value_counts()
```

```python
df.shape
```

Finding out the Null Values and Dropping them

```python
df.isna()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.shape
```

Re-Sampling the data 

```python

X = df.drop('Class', axis=1)
y = df['Class']

smt = SMOTE()
X_resampled, y_resampled = smt.fit_resample(X, y)
```

```python
X_resampled
```

```python
y_resampled
```

```python
y_resampled=pd.DataFrame(y_resampled)
```

```python
y_resampled
```

Re-Sampling successful through OVERSAMPLING technique 

```python
y_resampled.value_counts()
```

Scaling the data into a normalized form, using MIN ABS scaler function because we need values in range of -1 to 1  

```python
scaled=MaxAbsScaler()
```

```python
X_scaled=scaled.fit_transform(X_resampled)
```

```python
Scaled_df=pd.DataFrame(X_scaled,columns=['Time','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17',
                       'V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28','V29'])
```

Scaling successful hence proved by Statistical Measure of Data

```python
Scaled_df.describe().T
```

```python
Scaled_df['Target']=y_resampled
```

```python
Scaled_df
```

Data Sanitization and Engineering complete, saving the file into a new file named Feature_File.csv for Predictive Modelling.

```python
Scaled_df.to_csv('Feature_file.csv')
```
