```python
import pandas as pd
import xgboost as xgb
```

```python
df = pd.read_csv('inference_data.csv',index_col = 0)
df
```

```python
df.drop(columns=['Target','Time'],inplace=True)
df
```

```python
model = xgb.XGBClassifier()
model.load_model('XGBoost.json')
```

```python
prediction = model.predict(df)
for i in range(len(prediction)):
    if (prediction[i]== 0):
        print('The Transaction is not FRAUDULENT')
    else:
        print('The Transaction is FRAUDULENT')
```
